# Direct Methods for the solution of Linear Systems
<h4>Given a matrix A ∈ R n×n and the vector xtrue = (1, 1, . . . , 1)T ∈ R n, write a script that:
<ul>
<li>Computes the right-hand side of the linear system b = Axtrue.</li>
<li>Computes the condition number in 2-norm of the matrix A. It is ill-conditioned? What if we use
the ∞-norm instead of the 2-norm?</li>
<li>Solves the linear system Ax = b with the function np.linalg.solve().</li>
<li>Computes the relative error between the solution computed before and the true solution xtrue.</li>
<li>Plot a graph (using matplotlib.pyplot) with the relative errors as a function of n and (in a new window) the condition number in 2-norm K2(A) and in ∞-norm, as a function of n.</li>
</ul>
</h4>

```python
import numpy as np

# Setting up the dimension
n = 10
errors = []
# k2norms = []
# kinfnorms = []
kA2=[]
kAinf=[]
for c in range(1, n):
    # Creating the test problem
    A = np.random.randn(c, c)  # n x n random matrix
    x_true = np.ones((c, ))   # n-dimensional vector of ones

    y = A @ x_true  # Compute the term y s.t. x_true is a sol.

    # Solving the system with numpy
    x_sol = np.linalg.solve(A, y)
    KA2=np.linalg.cond(A,2)
    KAinf=np.linalg.cond(A,np.inf)
    # Computing the accuracy, calculating 2-norm and inf-norm also
    E_rel = np.linalg.norm(x_true - x_sol, 2) / np.linalg.norm(x_true, 2)
    norm_2 = np.linalg.norm(A, 2)
    norm_inf = np.linalg.norm(A, np.Inf)
    kA2.append(KA2)
    kAinf.append(KAinf)
    # k2norms.append(norm_2)
    # kinfnorms.append(norm_inf)
    errors.append(E_rel)
print(f"The relative error is {E_rel}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Plot a graph (using matplotlib.pyplot) with the relative errors as a function of n

# Visualize
plt.plot(errors)  # this opens the plot
plt.title('A plot of f(x) = relative error')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()

plt.plot(kA2)
plt.plot(kAinf)  # this opens the plot
plt.title('A plot of f(x) = condition numbers')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()

# and (in a newindow) the condition number in 2-norm K2(A) and in ∞-norm, as a function of n.

# Visualize
# plt.figure(figsize=(10, 4))
# # the matrix of 1 row and 2 columns put the plot in the first place
# plt.plot(k2norms, '-', color='red')
# plt.title('2-norms of K2(A)')
# plt.xlabel('x')
# plt.ylabel('y')
# plt.grid()
# plt.show()

# plt.plot(kinfnorms, '-', color='k')
# plt.title('∞-norm')
# plt.xlabel('x')
# plt.ylabel('y')
# plt.grid()

# plt.show()
```

## Vander Matrix

## Hilbert Matrix

```python
# 2. Test the program above with the following choices of A ∈ Rn×n:
# The Hilbert matrix (scipy.linalg.hilbert) of dimension n = {4, 5, 6, . . . , 12}
import numpy as np
from scipy.linalg import hilbert
import matplotlib.pyplot as plt
# Setting up the dimension

errors = []
k2norms = []
kinfnorms = []
kA2=[]
kAinf=[]
# setting the step of the for loop of 5 so I get the n I'm asked
for c in range(4, 12):

    A = hilbert(c)  # n x n random matrix
    x_true = np.ones((c, ))   # n-dimensional vector of ones

    y = A @ x_true  # Compute the term y s.t. x_true is a sol.

    # Solving the system with numpy
    x_sol = np.linalg.solve(A, y)
    KA2=np.linalg.cond(A,2)
    KAinf=np.linalg.cond(A,np.inf)

    # Computing the accuracy, calculating 2-norm and inf-norm also
    E_rel = np.linalg.norm(x_true - x_sol, 2) / np.linalg.norm(x_true, 2)
    norm_2 = np.linalg.norm(A, 2)
    norm_inf = np.linalg.norm(A, np.Inf)
    kA2.append(KA2)
    kAinf.append(KAinf)
    k2norms.append(norm_2)
    kinfnorms.append(norm_inf)
    errors.append(E_rel)
print(f"The relative error is {E_rel}")

# Visualize

x1=np.linspace(4,12,8,dtype=int)
plt.plot(x1,errors)  # this opens the plot
plt.title('A plot of f(x) = relative error')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()

plt.plot(x1,kA2)  # this opens the plot
plt.plot(x1,kAinf)
plt.title('A plot of f(x) = condition numbers')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(["kA2","kAinf"])
plt.grid()
plt.show()

# and (in a newindow) the condition number in 2-norm K2(A) and in ∞-norm, as a function of n.

# Visualize
# plt.figure(figsize=(10, 4))
# # the matrix of 1 row and 2 columns put the plot in the first place
# plt.subplot(1, 2, 1)
# plt.plot(k2norms, '-', color='red')
# plt.title('2-norms of K2(A)')
# plt.xlabel('x')
# plt.ylabel('y')
# plt.grid()

# plt.subplot(1, 2, 2)
# plt.plot(kinfnorms, '-', color='k')
# plt.title('∞-norm')
# plt.xlabel('x')
# plt.ylabel('y')
# plt.grid()

# plt.show()
```

# Floating point Arithmetic

is defined as the smallest floating point number such that it holds:


```python
epsilon = 1.0
counter = 1
while (1+epsilon) > 1:
    epsilon = epsilon/2
    counter += 1
print("value of epsilon is "+str(epsilon*2) +
      " reached after "+str(counter)+" loops")
print("if this command is reached it means the number got approximated to 0 hence the while loop ends")
```
